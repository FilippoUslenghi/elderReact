```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dummy_classifier.csv', header=[0,1], index_col=[0,1]).unstack(level=[0,1]).reset_index()
df.columns = ['features', 'pose', 'f1_score', 'emotion', 'value']
df
```

```python
fg = sns.catplot(x='features', y='value', hue='pose', col='emotion', data=(df[df.f1_score=='weighted_avg']), kind='bar', palette='deep')
fg.set_xlabels('')
plt.show()
```

```python
fg = sns.catplot(x='features', y='value', hue='pose', col='emotion', data=(df[df.f1_score=='macro_avg']), kind='bar', palette='deep')
fg.set_xlabels('')
plt.show()
```
